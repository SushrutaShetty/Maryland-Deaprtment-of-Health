## Trend Analysis

In this section we plot overall crime rate per 100,000 people over the years to see if there are any patterns or trends.

```python
##Year vs Population

year = crime_df['YEAR']
pop = crime_df['POPULATION']

fig, ax = plt.subplots(figsize =(16, 14))
ax.barh(year, pop)

for s in ['top','right']:
    ax.spines[s].set_visible(False)

ax.invert_yaxis()

## add annotations to bar
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')


plt.show()
```

```python
##Year vs Crime Rate per 100K people

year = crime_df['YEAR']
crime_rate = crime_df['OVERALL CRIME RATE PER 100K PEOPLE']

fig, ax = plt.subplots(figsize =(16, 14))
ax.barh(year, crime_rate)

for s in ['top','right']:
    ax.spines[s].set_visible(False)

ax.invert_yaxis()

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')


plt.show()
```

## Predictive Analysis

In this section we create a Machine Learning model to predict crime rate for year 2024 with a population of 8M. 



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#split the dataset into traing and test sets
train_data, test_data = train_test_split(crime_df, test_size = 0.2)
```

```python
# Input features and target variable
X_train = train_data[['POPULATION', 'YEAR']]
y_train = train_data['OVERALL CRIME RATE PER 100K PEOPLE']

X_test = test_data[['POPULATION', 'YEAR']]
y_test = test_data['OVERALL CRIME RATE PER 100K PEOPLE']


# machine learning algorithm
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squeared Error: ", mse)
print("R-squared: ", r2)

#Make pedictions on the new data
new_crime_rate = pd.DataFrame({'POPULATION': [8000000], 'YEAR':[2034]})

predicted_crime_rate = model.predict(new_crime_rate) 

print("Predicted crime rate for 2024: ", predicted_crime_rate)
```
